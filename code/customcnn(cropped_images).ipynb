```python
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# CELL 1 ── Install & verify GPU
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

!pip install -q tensorflow==2.15.0 scikit-learn matplotlib seaborn opencv-python-headless

import tensorflow as tf
from tensorflow.keras import mixed_precision
mixed_precision.set_global_policy("mixed_float16")

print("TensorFlow :", tf.__version__)
print("GPU        :", tf.config.list_physical_devices("GPU"))
print("Policy     :", mixed_precision.global_policy().name)


# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# CELL 2 ── Mount Drive & Configuration
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

from google.colab import drive
drive.mount("/content/drive")

import os, math, warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
from pathlib import Path
from sklearn.metrics import (
    confusion_matrix, classification_report, roc_auc_score,
    roc_curve, f1_score, average_precision_score,
    precision_recall_curve, matthews_corrcoef
)
from tensorflow.keras import layers, Model, callbacks, optimizers, regularizers

warnings.filterwarnings("ignore")
tf.get_logger().setLevel("ERROR")

# ── Paths ─────────────────────────────────────────────────────────────────
SPLIT_DIR = Path("/content/drive/MyDrive/Breast_cancer_dataset/cropped_split")
SAVE_DIR  = Path("/content/drive/MyDrive/Breast_cancer_dataset/model_output/custom_cnn")
SAVE_DIR.mkdir(parents=True, exist_ok=True)

TRAIN_DIR = SPLIT_DIR / "train"
VAL_DIR   = SPLIT_DIR / "val"
TEST_DIR  = SPLIT_DIR / "test"

# ── Settings ───────────────────────────────────────────────────────────────
MODEL_NAME   = "CustomCNN"
IMG_SIZE     = 224
CHANNELS     = 3
BATCH_SIZE   = 32
TOTAL_EPOCHS = 50
INIT_LR      = 1e-3
DROPOUT_RATE = 0.5
L2_REG       = 1e-4
RANDOM_SEED  = 42

tf.random.set_seed(RANDOM_SEED)
np.random.seed(RANDOM_SEED)

print("=" * 65)
print(f"  Model      : {MODEL_NAME}  (no pretrained weights)")
print(f"  Input size : {IMG_SIZE}×{IMG_SIZE}×{CHANNELS}")
print(f"  Epochs     : {TOTAL_EPOCHS}  (single-phase)")
print(f"  Save dir   : {SAVE_DIR}")
print("=" * 65)


# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# CELL 3 ── Dataset inspection & class weights
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

IMG_EXTS = (".jpg", ".jpeg", ".png")
stats = {}
for split in ["train", "val", "test"]:
    counts = {}
    for label in ["malignant", "benign"]:
        p = SPLIT_DIR / split / label
        counts[label] = len([f for f in p.iterdir()
                             if f.suffix.lower() in IMG_EXTS]) if p.exists() else 0
    stats[split] = counts

print(f"  {'Split':<8} {'Malignant':>10} {'Benign':>8} {'Total':>7}  Mal%")
print(f"  {'─'*42}")
for split, cnt in stats.items():
    mal = cnt["malignant"]; ben = cnt["benign"]; tot = mal + ben
    print(f"  {split:<8} {mal:>10,} {ben:>8,} {tot:>7,}  {mal/tot*100:.1f}%")

n_mal = stats["train"]["malignant"]
n_ben = stats["train"]["benign"]
n_tot = n_mal + n_ben
CLASS_WEIGHTS = {0: n_tot/(2*n_ben), 1: n_tot/(2*n_mal)}
print(f"\n  Class weights → benign:{CLASS_WEIGHTS[0]:.4f}  malignant:{CLASS_WEIGHTS[1]:.4f}")


# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# CELL 4 ── Preprocessing & augmentation
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

def apply_clahe(img_np):
    img_u8 = img_np.astype(np.uint8)
    lab     = cv2.cvtColor(img_u8, cv2.COLOR_RGB2LAB)
    l, a, b = cv2.split(lab)
    clahe   = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    lab_eq  = cv2.merge([clahe.apply(l), a, b])
    return cv2.cvtColor(lab_eq, cv2.COLOR_LAB2RGB)

def preprocess_train(image_path, label):
    raw   = tf.io.read_file(image_path)
    image = tf.image.decode_jpeg(raw, channels=CHANNELS)
    image = tf.image.resize(image, [IMG_SIZE, IMG_SIZE])
    image = tf.cast(image, tf.uint8)
    image = tf.py_function(lambda x: apply_clahe(x.numpy()), [image], tf.uint8)
    image.set_shape([IMG_SIZE, IMG_SIZE, 3])
    # Custom CNN: simple [0, 1] normalisation (no ImageNet stats)
    image = tf.cast(image, tf.float32) / 255.0
    image = tf.image.random_flip_left_right(image)
    image = tf.image.random_flip_up_down(image)
    crop_pct  = tf.random.uniform([], 0.85, 1.0)
    crop_size = tf.cast(tf.cast(IMG_SIZE, tf.float32) * crop_pct, tf.int32)
    image = tf.image.random_crop(image, [crop_size, crop_size, 3])
    image = tf.image.resize(image, [IMG_SIZE, IMG_SIZE])
    image = tf.image.random_brightness(image, max_delta=0.2)
    image = tf.image.random_contrast(image, lower=0.8, upper=1.2)
    noise = tf.random.normal(tf.shape(image), mean=0.0, stddev=0.02)
    image = tf.clip_by_value(image + noise, 0.0, 1.0)
    return image, tf.cast(label, tf.float32)

def preprocess_val(image_path, label):
    raw   = tf.io.read_file(image_path)
    image = tf.image.decode_jpeg(raw, channels=CHANNELS)
    image = tf.image.resize(image, [IMG_SIZE, IMG_SIZE])
    image = tf.cast(image, tf.uint8)
    image = tf.py_function(lambda x: apply_clahe(x.numpy()), [image], tf.uint8)
    image.set_shape([IMG_SIZE, IMG_SIZE, 3])
    image = tf.cast(image, tf.float32) / 255.0
    return image, tf.cast(label, tf.float32)


# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# CELL 5 ── tf.data pipelines
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

def collect(split_path):
    paths, labels = [], []
    for name, val in [("malignant",1),("benign",0)]:
        folder = split_path / name
        if folder.exists():
            for f in sorted(folder.iterdir()):
                if f.suffix.lower() in IMG_EXTS:
                    paths.append(str(f)); labels.append(val)
    return paths, labels

def make_ds(split_path, fn, shuffle=False):
    paths, labels = collect(split_path)
    ds = tf.data.Dataset.from_tensor_slices(
        (tf.constant(paths), tf.constant(labels, dtype=tf.int32)))
    if shuffle:
        ds = ds.shuffle(len(paths), seed=RANDOM_SEED, reshuffle_each_iteration=True)
    return (ds.map(fn, num_parallel_calls=tf.data.AUTOTUNE)
              .batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE),
            len(paths), paths, labels)

train_ds, n_train, _, train_labels = make_ds(TRAIN_DIR, preprocess_train, True)
val_ds,   n_val,   _, val_labels   = make_ds(VAL_DIR,   preprocess_val)
test_ds,  n_test,  _, test_labels  = make_ds(TEST_DIR,  preprocess_val)
steps_per_epoch = math.ceil(n_train / BATCH_SIZE)
print(f"  Train:{n_train:,}  Val:{n_val:,}  Test:{n_test:,}  Steps/ep:{steps_per_epoch}")


# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# CELL 6 ── Build Custom CNN
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

print(f"\n▶  Building {MODEL_NAME} …\n")

inputs = layers.Input(shape=(IMG_SIZE, IMG_SIZE, CHANNELS), name="input")
x = inputs

# 4 convolutional blocks — increasing depth (32 → 64 → 128 → 256)
for filters, name in [(32,"b1"),(64,"b2"),(128,"b3"),(256,"b4")]:
    x = layers.Conv2D(filters, 3, padding="same",
                      kernel_regularizer=regularizers.l2(L2_REG),
                      name=f"{name}_conv1")(x)
    x = layers.BatchNormalization(name=f"{name}_bn1")(x)
    x = layers.ReLU(name=f"{name}_relu1")(x)
    x = layers.Conv2D(filters, 3, padding="same",
                      kernel_regularizer=regularizers.l2(L2_REG),
                      name=f"{name}_conv2")(x)
    x = layers.BatchNormalization(name=f"{name}_bn2")(x)
    x = layers.ReLU(name=f"{name}_relu2")(x)
    x = layers.MaxPooling2D(2, name=f"{name}_pool")(x)
    x = layers.Dropout(0.25, name=f"{name}_drop")(x)

# Classification head
x   = layers.GlobalAveragePooling2D(name="gap")(x)
x   = layers.Dense(512, activation="relu",
                   kernel_regularizer=regularizers.l2(L2_REG), name="fc1")(x)
x   = layers.BatchNormalization(name="bn1")(x)
x   = layers.Dropout(DROPOUT_RATE, name="drop1")(x)
x   = layers.Dense(256, activation="relu",
                   kernel_regularizer=regularizers.l2(L2_REG), name="fc2")(x)
x   = layers.BatchNormalization(name="bn2")(x)
x   = layers.Dropout(0.3, name="drop2")(x)
out = layers.Dense(1, activation="sigmoid", dtype="float32", name="output")(x)

model = Model(inputs, out, name="CustomCNN_BreastCancer")
model.summary(line_length=80)
total_params = sum(tf.size(w).numpy() for w in model.weights)
print(f"  Total params : {total_params:,}  (all trainable — no frozen backbone)")


# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# CELL 7 ── Cosine LR schedule
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

def cosine_schedule(init_lr, total_epochs, steps_ep, warmup_ep=3, min_lr=1e-7):
    cosine = tf.keras.optimizers.schedules.CosineDecayRestarts(
        initial_learning_rate=init_lr,
        first_decay_steps=max((total_epochs - warmup_ep) * steps_ep // 3, 1),
        t_mul=2.0, m_mul=0.9, alpha=min_lr/init_lr)
    warmup_steps = warmup_ep * steps_ep
    class Schedule(tf.keras.optimizers.schedules.LearningRateSchedule):
        def __call__(self, step):
            step = tf.cast(step, tf.float32)
            return tf.where(step < warmup_steps,
                            init_lr * (step / max(warmup_steps, 1)),
                            cosine(tf.maximum(step - warmup_steps, 0)))
        def get_config(self): return {}
    return Schedule()

METRICS = [
    "accuracy",
    tf.keras.metrics.AUC(name="auc",    curve="ROC"),
    tf.keras.metrics.AUC(name="pr_auc", curve="PR"),
    tf.keras.metrics.Precision(name="precision"),
    tf.keras.metrics.Recall(name="recall"),
]


# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# CELL 8 ── Training  (single phase — no backbone to unfreeze)
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

print(f"\n▶  Training {MODEL_NAME} from scratch ({TOTAL_EPOCHS} epochs) …\n")
best_ckpt = str(SAVE_DIR / "best.weights.h5")

model.compile(
    optimizer=optimizers.Adam(cosine_schedule(INIT_LR, TOTAL_EPOCHS, steps_per_epoch)),
    loss="binary_crossentropy",
    metrics=METRICS,
)

history = model.fit(
    train_ds,
    epochs          = TOTAL_EPOCHS,
    validation_data = val_ds,
    class_weight    = CLASS_WEIGHTS,
    verbose         = 1,
    callbacks=[
        callbacks.ModelCheckpoint(best_ckpt, monitor="val_auc", mode="max",
                                  save_best_only=True, save_weights_only=True, verbose=1),
        callbacks.EarlyStopping(monitor="val_auc", patience=12, mode="max",
                                restore_best_weights=True, verbose=1),
        callbacks.TensorBoard(log_dir=str(SAVE_DIR/"logs")),
        callbacks.CSVLogger(str(SAVE_DIR/"training_history.csv")),
    ],
)

model.save(str(SAVE_DIR / "custom_cnn_breast_cancer.keras"))
print(f"  ✔ Training done. Best val_AUC = {max(history.history['val_auc']):.4f}")


# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# CELL 9 ── Training history plot
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle(f"{MODEL_NAME} Training History", fontsize=14, fontweight="bold")
for ax, (tk, vk, title) in zip(axes.flat, [
    ("loss","val_loss","Loss"),("accuracy","val_accuracy","Accuracy"),
    ("auc","val_auc","ROC-AUC"),("pr_auc","val_pr_auc","PR-AUC"),
    ("precision","val_precision","Precision"),("recall","val_recall","Recall"),
]):
    ep = range(1, len(history.history[tk])+1)
    ax.plot(ep, history.history[tk], "b-o", ms=3, label="Train")
    ax.plot(ep, history.history[vk], "r-o", ms=3, label="Val")
    ax.set_title(title); ax.legend(fontsize=8); ax.grid(alpha=0.3)
plt.tight_layout()
plt.savefig(str(SAVE_DIR/"training_history.png"), dpi=150, bbox_inches="tight")
plt.show()


# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# CELL 10 ── Evaluation
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

print("\n▶  Test set evaluation …\n")
model.load_weights(best_ckpt)

y_prob = model.predict(test_ds, verbose=1).flatten()
y_true = np.array(test_labels)
y_pred = (y_prob >= 0.5).astype(int)

fpr_a, tpr_a, thr_a = roc_curve(y_true, y_prob)
opt_idx    = np.argmax(tpr_a - fpr_a)
opt_thresh = float(thr_a[opt_idx])
auc_roc    = roc_auc_score(y_true, y_prob)
ap         = average_precision_score(y_true, y_prob)
cm         = confusion_matrix(y_true, y_pred)
tn, fp, fn, tp = cm.ravel()

accuracy    = (tp+tn)/(tp+tn+fp+fn)
sensitivity = tp/(tp+fn)
specificity = tn/(tn+fp)
precision_v = tp/(tp+fp) if (tp+fp)>0 else 0
npv         = tn/(tn+fn) if (tn+fn)>0 else 0
f1          = f1_score(y_true, y_pred)
mcc         = matthews_corrcoef(y_true, y_pred)

print(f"  Accuracy:{accuracy:.4f}  AUC:{auc_roc:.4f}  F1:{f1:.4f}")
print(f"  Sensitivity:{sensitivity:.4f}  Specificity:{specificity:.4f}")
print(f"  TP={tp} TN={tn} FP={fp} FN={fn}  OptThr={opt_thresh:.4f}")
print()
print(classification_report(y_true, y_pred, target_names=["benign","malignant"]))

fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle(f"{MODEL_NAME} — Test Set Evaluation", fontsize=13, fontweight="bold")
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", ax=axes[0],
            xticklabels=["Pred Benign","Pred Malignant"],
            yticklabels=["True Benign","True Malignant"], annot_kws={"size":14})
axes[0].set_title("Confusion Matrix (thr=0.50)")
axes[1].plot(fpr_a, tpr_a, "b-", lw=2, label=f"AUC={auc_roc:.4f}")
axes[1].plot([0,1],[0,1],"k--")
axes[1].scatter(fpr_a[opt_idx], tpr_a[opt_idx], color="red", s=80, label=f"Opt={opt_thresh:.2f}")
axes[1].set_xlabel("FPR"); axes[1].set_ylabel("TPR"); axes[1].set_title("ROC Curve")
axes[1].legend(); axes[1].grid(alpha=0.3)
pr_p, pr_r, _ = precision_recall_curve(y_true, y_prob)
axes[2].plot(pr_r, pr_p, "g-", lw=2, label=f"AP={ap:.4f}")
axes[2].axhline(y_true.mean(), color="r", ls="--")
axes[2].set_xlabel("Recall"); axes[2].set_ylabel("Precision"); axes[2].set_title("PR Curve")
axes[2].legend(); axes[2].grid(alpha=0.3)
plt.tight_layout()
plt.savefig(str(SAVE_DIR/"evaluation_plots.png"), dpi=150, bbox_inches="tight"); plt.show()

results = {
    "model":"custom_cnn","img_size":IMG_SIZE,
    "accuracy":round(accuracy,4),"auc_roc":round(auc_roc,4),"pr_auc":round(ap,4),
    "f1":round(f1,4),"mcc":round(mcc,4),"sensitivity":round(sensitivity,4),
    "specificity":round(specificity,4),"precision":round(precision_v,4),"npv":round(npv,4),
    "opt_threshold":round(opt_thresh,4),
    "TP":int(tp),"TN":int(tn),"FP":int(fp),"FN":int(fn),"miss_rate":round(fn/(fn+tp),4),
}
pd.DataFrame([results]).to_csv(str(SAVE_DIR/"results_summary.csv"), index=False)
master_csv = str(SAVE_DIR.parent/"all_models_comparison.csv")
df = pd.read_csv(master_csv) if os.path.exists(master_csv) else pd.DataFrame()
df = df[df["model"] != "custom_cnn"] if len(df) else df
df = pd.concat([df, pd.DataFrame([results])], ignore_index=True)
df.to_csv(master_csv, index=False)
print(f"\n  ✅  CustomCNN complete!  AUC={auc_roc:.4f}  Sens={sensitivity:.4f}  F1={f1:.4f}")
```
